```python
# first target to seperate date column and message column
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
# this pattern is for 24hours format 
# to build regular expression we can use reges1o1.com (paste message over there we will get regulr expression)
```

```python
# messages=re.split(pattern,data) 
# messages
```

```python
# somehow at strat we are getting 1 empty string to remove that
messages=re.split(pattern,data)[1:] 
messages
```

```python
dates=re.findall(pattern,data) 
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})

# Converting 'message_date' type
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %H:%M - ')

# Renaming the 'message_date' column to 'date'
df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
# to see how many messages are there
df.shape
```

```python
# seperating users and messages
users=[]
messages=[]
for message in df['user_message']:
    entry=re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['user'] =users
df['message']=messages
df.drop(columns=['user_message'],inplace=True)

df.head()
```

```python
# here in hours column if it shows 12 and some minutes in minute column so we can say that lies between 12-13 in hours
```

```python
period=[]
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str ('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period']=period
```

```python
df.head()
```

```python
import seaborn as sns
plt.figure(figsize=(20,6))
sns.heatmap(df.pivot_table(index='day_name', columns='period',values= 'message', aggfunc='count').fillna(0))
plt.yticks(rotation= 'horizontal')
plt.show()
```
